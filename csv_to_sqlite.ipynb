```python
import pandas as pd
import sqlite3

# Chemin du fichier CSV
csv_file = "C:/Users/AZIMUT Info/OneDrive/Desktop/repos/data_spiro.csv"  # Remplacez par le chemin vers votre fichier CSV

# Charger le CSV avec pandas
df = pd.read_csv(csv_file)

# Afficher les premières lignes du CSV pour vérification
print("Aperçu des données :")
print(df.head())

# Connexion à une base de données SQLite (créera le fichier si nécessaire)
conn = sqlite3.connect("ma_base_de_donnees.db")
cursor = conn.cursor()

# Création de la table (selon la structure du CSV)
table_name = "ma_table"

# Générer une commande SQL CREATE TABLE dynamiquement à partir des colonnes du DataFrame
columns = ", ".join([f"{col} TEXT" for col in df.columns])  # On utilise TEXT par défaut
create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns});"

# Exécuter la commande de création de table
cursor.execute(create_table_query)

# Insérer les données du DataFrame dans la table
df.to_sql(table_name, conn, if_exists='append', index=False)

# Vérifier si les données ont été correctement insérées
cursor.execute(f"SELECT * FROM {table_name} LIMIT 5;")
rows = cursor.fetchall()
print("Aperçu des données insérées :")
for row in rows:
    print(row)

# Fermer la connexion
conn.close()
print("Base de données créée avec succès.")
```
